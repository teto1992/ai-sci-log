```python
import pandas as pd
from random import randint
```

```python
a=pd.read_csv('regional_carbon_intensity_2025_complete.csv')
a.head()
```

```python
# only keep numerical data
b=a[['regionid', 'forecast', 'datetime']]
# read in the data range
b.describe()
```

```python
# randomised CPU%
b['CPU(%)']=randint(1,100)
# corresponding energy usagage for 10 hours
b['Energy (kWh, 10h)']=b['CPU(%)']*0.003+0.2
# carbon emission using the regional carbon intensity forecast data
b['gCO2']=b['Energy (kWh, 10h)']*b['forecast']
# reserved hours = gCO2 emission /10 
b['Reserved_Hour_per_CO2']=b['gCO2']/10 
b['gCO2_per_Active_Compute_Hour']= b['Reserved_Hour_per_CO2']/b['CPU(%)']/100
b.head()
```

```python
b.to_csv('data1_without_FU.csv')
```
